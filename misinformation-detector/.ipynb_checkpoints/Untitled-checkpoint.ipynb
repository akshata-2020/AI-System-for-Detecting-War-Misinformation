```python
# Cell 1 — All imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, learning_curve
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    precision_score, recall_score, f1_score
)

import joblib

print("All imports successful ✅")
```

```python
fake = pd.read_csv(r'C:\Users\DELL\misinformation-detector\data\Fake.csv')
real = pd.read_csv(r'C:\Users\DELL\misinformation-detector\data\True.csv')
```

```python
# Add labels
fake['label'] = 0  # 0 = Fake
real['label'] = 1  # 1 = Real

# Combine
df = pd.concat([fake, real], ignore_index=True)

# ── Always look at your data first (Ng's rule) ──
print(df.shape)
print(df.head())
print(df.isnull().sum())

# Check class balance
df['label'].value_counts().plot(kind='bar', color=['red','green'])
plt.title('Fake vs Real Count')
plt.xticks([0,1], ['Fake','Real'], rotation=0)
plt.show()

# Headline length distribution
df['title_length'] = df['title'].apply(lambda x: len(x.split()))
df.groupby('label')['title_length'].hist(alpha=0.6, bins=30)
plt.title('Headline Word Count by Label')
plt.legend(['Fake','Real'])
```
